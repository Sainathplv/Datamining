```r
# Read the file
df_data <- read.csv("healthcarestrokedataset.csv", header = TRUE)
```

```r
#return the first 5 rows of the dataset
head(df_data,5)
```

```r
#Data Pre-processing for R
# Explain each step in detail with reason

#1. To check the number of Null in the data set
#The first step after loading the data to R would be to check for possible issues 
#such as missing data, outliers, and so on, and, depending on the analysis, the preprocessing operation will be decided. 
#Usually, in any dataset, the missing values have to be dealt with either by not 
#considering them for the analysis or replacing them with a suitable value.

sum(is.na(df_data))

#2. Replace the missing data with the average of the feature in which the data is missing
#This is to deal with the missing values
df_data$age = ifelse(is.na(df_data$age),ave(df_data$age, FUN = function (x)mean(x, na.rm = TRUE)),df_data$age)
df_data

#3. Encoding categorical data
#Encoding refers to transforming text data into numeric data. 
#Encoding Categorical data simply means we are transforming data that fall into categories into numeric data.

df_data$gender = factor(df_data$gender, levels = c('Male','Female'), labels = c(1.0, 2.0 ))
df_data

#4. Splitting the dataset into the training and test set
#The reason is that when the dataset is split into train and test sets, 
#there will not be enough data in the training dataset for the model to learn an effective mapping of inputs to outputs.
set.seed(123)
split = sample.split(df_data$smoking_status, SplitRatio = 0.8)# returns true if observation goes to the Training set and false if observation goes to the test set.

#Creating the training set and test set separately
training_set = subset(df_data, split == TRUE)
test_set = subset(df_data, split == FALSE)
training_set
test_set

#5. Feature scaling
#Feature scaling is the process of eliminating units of measurement for variables within a dataset                     
training_set[, 2:3] = scale(training_set[, 2:3])
test_set[, 2:3] = scale(test_set[, 2:3])
training_set
test_set                   
```
